```python
import pandas as pd
import numpy as np


# Step 1: Create the 'datetime' column by combining Year, Month, and Day
df['datetime'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str) + '-' + df['Day'].astype(str))

# Step 2: Extract features based on the 'datetime' column
df['day_of_year'] = df['datetime'].dt.dayofyear
df['week_of_year'] = df['datetime'].dt.isocalendar().week
df['quarter'] = df['datetime'].dt.quarter
df['weekday'] = df['datetime'].dt.weekday
df['is_weekend'] = (df['weekday'] >= 5).astype(int)

# Step 3: Encode cyclic features (for month and day of the year)
df['month_sin'] = np.sin(2 * np.pi * df['Month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['Month'] / 12)

df['day_sin'] = np.sin(2 * np.pi * df['day_of_year'] / 365)
df['day_cos'] = np.cos(2 * np.pi * df['day_of_year'] / 365)

# Step 4: Create lag features (using past temperature as an example)
df['temp_lag_1'] = df['Avg_Temperature'].shift(1)  # 1-day lag for temperature

# Step 5: Remove the 'Year', 'Month', and 'Day' columns
df.drop(['Year', 'Month', 'Day'], axis=1, inplace=True)

# Step 6: Reorder columns to place time-related features after the 'ID' column
time_columns = ['month_sin', 'month_cos', 'day_sin', 'day_cos', 'week_of_year', 'quarter', 'weekday', 'is_weekend', 'temp_lag_1']
df = df[['ID'] + time_columns + [col for col in df.columns if col not in time_columns + ['ID']]]

# Step 7: Save the updated DataFrame
df.to_csv('weather_prediction_data.csv', index=False)

# Display the final DataFrame
print(df)
```

```python
import pandas as pd
import numpy as np

# Load the CSV file
file_path = r"C:\Users\pirag\Documents\Data_Crunch_Competition\modified_temperatures.csv"
df = pd.read_csv(file_path)

# Ensure Year, Month, Day columns exist
if {'Year', 'Month', 'Day'}.issubset(df.columns):
    # Create cyclic features for month
    df['Month_sin'] = np.sin(2 * np.pi * df['Month'] / 12)
    df['Month_cos'] = np.cos(2 * np.pi * df['Month'] / 12)

    # Create cyclic features for day
    df['Day_sin'] = np.sin(2 * np.pi * df['Day'] / 31)
    df['Day_cos'] = np.cos(2 * np.pi * df['Day'] / 31)

    # Remove original Month and Day columns
    df.drop(columns=['Month', 'Day'], inplace=True)

    # Reorder columns: Year first, then cyclic features, then other columns
    cols = ['Year', 'Month_sin', 'Month_cos', 'Day_sin', 'Day_cos'] + [col for col in df.columns if col not in ['Year', 'Month_sin', 'Month_cos', 'Day_sin', 'Day_cos']]
    df = df[cols]

    # Save the updated DataFrame
    updated_file_path = file_path.replace(".csv", "_updated.csv")
    df.to_csv(updated_file_path, index=False)

    print(f"âœ… Updated CSV saved as: {updated_file_path}")
else:
    print("âŒ Error: 'Year', 'Month', or 'Day' columns not found in the CSV file.")
```
